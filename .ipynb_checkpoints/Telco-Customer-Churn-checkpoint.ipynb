<b>Phase 1: Data Preprocessing & Feature Engineering</b>

1.1: Extract & Initial Exploration

```python
import pandas as pd
```

```python
df = pd.read_csv("Telco-Customer-Churn.csv")
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

1.2: Data Cleaning and Formatting

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
print(df.isnull().sum())
```

```python
df.drop(columns = 'customerID', inplace=True)
```

```python
print(df.info())
```

```python
df.to_csv('cleaned_telco-customer-churn.csv', index = False)
```

```python
df['Churn'] = df['Churn'].apply(lambda x:1 if x == 'Yes' else 0)
```

```python
print(df.info())
```

1.3: One-Hot Encoding for Categorical Features

```python
X = df.drop('Churn', axis = 1)
Y = df['Churn']
```

```python
X_encoded = pd.get_dummies(X,drop_first = True)
```

```python
print(X_encoded.shape)
```

```python
print(X_encoded.info())
```

```python
print(X_encoded.head())
```

<b>Phase 2: Building and Evaluating the ML Model</b>

```python
from sklearn.model_selection import train_test_split
```

2.1: Splitting the Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, Y, test_size=0.2, random_state=42)
```

2.2: Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
print(X_train_scaled[:5])
```

```python
X_test_scaled = scaler.transform(X_test)
```

2.3: Training the Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter = 1000)
model.fit(X_train_scaled, Y_train)
```

2.4: Making Predictions and Evaluating Performance

```python
y_pred = model.predict(X_test_scaled)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

accuracy = accuracy_score(Y_test, y_pred)
print(f"\nModel Accuracy: {accuracy:.4f}") 

print("\n--- Confusion Matrix ---")
cm = confusion_matrix(Y_test, y_pred)
print(cm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Churn', 'Churn'], yticklabels=['Not Churn', 'Churn'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.savefig('confusion_matrix.png') 
plt.show()

print("\n--- Classification Report ---")
print(classification_report(Y_test, y_pred))
```
